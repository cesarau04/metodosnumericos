# Realizando equaciones a la antigua
Vamos a realizar la siguiente ecuacion  
$ x^3 - 7x^2 + 8x + 0.35 $

```python
def y(x):
    return x**3 - 7*x**2 + 8*x + 0.35

print(y(1.37))
```

# Analitico vs Num√©rico (Metodo Grafico)
$ ax^2 + bx + c = 0$

Analitico:
$y = \frac{-b \pm \sqrt[\leftroot{-2}\uproot{2}]{b^2 - 4ac}}{2a}$

Grafico:
Darle valor a $x$ y graficar
$2x^2 - 5x + 3 = 0$

```python
import numpy as np
import matplotlib.pyplot as plt

def y(x):
    return 1-((20**2/9.81)*(3+x)/(3*x+((x**2)/2))**3)

xarray = np.linspace(1, 4,100)

yarray = np.zeros(100)

for i in range(len(xarray)):
    yarray[i] = y(xarray[i])

plt.plot(xarray, yarray)
plt.grid()
plt.show()
```
