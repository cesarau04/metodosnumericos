# Metodo de falsa posicion
  
Elige la x mas cercana a 0 para poder realizar el proceso mas rapido

$ x_r = x_0 + \frac{ f_{x_0}(x_1 - x_0)}{(f_{x_1} - f_{x_0}} $

###### Intentar programarlo
1. Ademas de checar la resolucion

```python
# Funcion para problema de paracaidista
import math
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    #return (667.38/c)*(1-math.exp(-6.81*c))-40
    #return x**2 - 10
    return 1-((20**2/9.81)*(3+x)/(3*x+((x**2)/2))**3)

x = np.linspace(1,2,100)
xa = 1
xb = 2
xc = ((xa*f(xb))-(xb*f(xa)))/(f(xb)-f(xa))

while abs(f(xc)) > 0.005:
    if f(xc) * f(xa) > 0:
         xa = xc
    else:
        xb = xc
    xc = ((xa * f(xb)) - (xb * f(xa))) / (f(xb) - f(xa))

print("The root is",xc)
plt.plot(x,f(x))
plt.grid(True)
plt.show()
```
